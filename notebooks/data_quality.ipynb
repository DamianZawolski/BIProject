```python
import os
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def analyze_csv(df):
    
    # Get the column names
    columns = df.columns.tolist()
    
    # Create a dictionary to store the analysis results
    analysis_results = {}
    
    # Iterate over each column
    for col in columns:
        # Get distinct values
        distinct_values = df[col].dropna().nunique()
        
        # Get unique values
        unique_values = df[col].dropna().value_counts().loc[lambda x: x == 1].count()
        
        # Get null values count
        null_values_count = df[col].isnull().sum()
        
        
        # Store the analysis results for the current column
        analysis_results[col] = {
            'distinct_values': distinct_values,
            'unique_values': unique_values,
            'null_values_count': null_values_count,
        }
    
    return analysis_results
```

```python
csv_file = 'tripdata_connected.csv'
df = pd.read_csv(csv_file)
results = analyze_csv(df)
```

```python
print(f'Overall values count: {len(df)}')
for column, analysis in results.items():
    print(f"Column: {column}")
    print(f"Duplicated Values: {analysis['distinct_values']- analysis['unique_values']}")
    print(f"Unique Values: {analysis['unique_values']}")
    print(f"Null Values Count: {analysis['null_values_count']}")
    print()
```

```python
csv_file = 'weather_data.csv'
df = pd.read_csv(csv_file, encoding="Windows-1250", sep=";")
results = analyze_csv(df)
```
