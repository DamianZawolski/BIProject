Importing data from CSV

```python
import pandas as pd
df = pd.read_csv(r'Spotify_Youtube.csv')
df.head()
```

```python
df.columns
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
fig = plt.figure(figsize=(20, 15))
x = df.Views
y = df.Danceability
plt.scatter(x, y, color="red")

plt.title("Relation between Dancebility and Youtube views")
plt.xlabel("YT Views")
plt.ylabel(f"Dancebility")
plt.show()
```

```python
fig.clear()
fig = plt.figure(figsize=(20, 15))
sns.set_theme()
x = df.Stream
y = df.Danceability
plt.scatter(x, y, color="green")
plt.title("Relation between Dancebility and Spotify streams")
plt.xlabel("Spotify streams")
plt.ylabel(f"Dancebility")
plt.show()
```

```python
fig.clear()
fig = plt.figure(figsize=(20, 15))
sns.set_theme()
x = df.Stream
y = df.Views
plt.scatter(x, y, color="black")
plt.title("Relation between YT Views and Spotify streams")
plt.xlabel(f"Spotify Streams")
plt.ylabel("YT Views")
plt.show()
```
