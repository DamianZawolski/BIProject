Importing data from CSV

```python
import pandas as pd
df = pd.read_csv(r'Spotify_Youtube.csv')
print(df.head())
```

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20, 15))
x = df.Views
y = df.Danceability
plt.scatter(x, y, color="red")
plt.title("Relation between Dancebility and Youtube views")
plt.xlabel("YT Views")
plt.ylabel(f"Dancebility")
plt.show()
```

```python
fig.clear()
fig = plt.figure(figsize=(20, 15))
x = df.Stream
y = df.Danceability
plt.scatter(x, y, color="green")
plt.title("Relation between Dancebility and Spotify streams")
plt.xlabel("YT Views")
plt.ylabel(f"Dancebility")
plt.show()
```
